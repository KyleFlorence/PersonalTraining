```python
from datetime import date
```

```python
def load_dict_from_file():
  try:
    f = open('personal_training_dict.txt', 'r')
    data=f.read()
    f.close()
    return eval(data)
  except:
    f = open('personal_training_dict.txt', 'w')
    f.close()
    return {}

db = load_dict_from_file()  
```

```python
def new_exercise(name, setsreps, weight):
  exercise = {
      "exercise_name": name,
      "sets x reps": setsreps,
      "weight": weight 
  }
  return exercise

def add_client():
  name = input("What is the client\'s name? \n")
  if name.lower() in db:
    print("Client already in database.")
    print("Performing this action will overwrite the existing client.")
    answer = input("Enter 1 to confirm, enter anything else to cancel.\n")
    if answer == "1":
      db[name.lower()] = {}
  else:
    db[name.lower()] = {}

def check_exercise(exercise, client_name):
  for workout in reversed(list(db[client_name].keys())):
    for lift in range(len(db[client_name][workout])):
      if db[client_name][workout][lift]['exercise_name'] == exercise:
        return "Last performed on {}: {}, for {}".format(workout, db[client_name][workout][lift]['weight'], db[client_name][workout][lift]['sets x reps'])
  return "Exercise not found in client\'s history."

def ask_for_exercise(list_of_exercises, client_name):
  exercise = input("Enter an exercise name, or type 'cancel' to cancel. \n")
  
  if exercise != 'cancel':
    # display last time it was performed
    print(check_exercise(exercise.lower(), client_name.lower()))
    setsreps = input("How many sets and reps were performed? \n")
    weight = input("How much weight was lifted? \n")
    new_exercise_dict = new_exercise(exercise.lower(), setsreps, weight)
    list_of_exercises.append(new_exercise_dict)

def start_workout(client_name):
  exercises = []
  print("Workout started! \n")
  while True:
    ask_for_exercise(exercises, client_name)
    print()
    print("Want to add another exercise?")
    answer = input("Enter 1 to add another, enter anything else to quit workout \n")
    print()
    if answer != "1":
      break
  return exercises

def new_workout(client_name):
  exercises = start_workout(client_name)
  db[client_name][date.today().strftime('%m/%d/%Y')] = exercises

def view_workout():
  client_name = input("Enter the client\'s name. \n")
  if client_name.lower() not in db:
    print("Client not found in database.")
  else:
    workout_date = input("Enter the date of the workout (mm/dd/yyyy). \n")
    if workout_date not in db[client_name.lower()]:
      print("No record found on that date.")
    else:
      for lift in range(len(db[client_name.lower()][workout_date])):
        print(db[client_name.lower()][workout_date][lift]['exercise_name'] + ", " + db[client_name.lower()][workout_date][lift]['sets x reps'] + ", " + db[client_name.lower()][workout_date][lift]['weight'])

def view_clients():
  print()
  print("Client names:")
  for key in db.keys():
    print(key.title())

def save_dict_to_file(dic):
    f = open('personal_training_dict.txt', 'w')
    f.write(str(dic))
    f.close()

def main():
  print("Welcome to the gym!")

  while True:
    print()
    print("The options are as follows:")
    print("1: Add a client")
    print("2: Start a workout for existing client")
    print("3: View a previous workout for an existing client")
    print("4: View all client names")
    print()
    answer = input("Choose 1-4. Enter anything else to quit. \n")

    if answer == "1":
      add_client()
    elif answer == "2":
      client_name = input("which client is working out?\n")
      if client_name.lower() in db:
        new_workout(client_name.lower())
      else:
        print("Client not found in database.")
    elif answer == "3":
      view_workout()
    elif answer == "4":
      view_clients()
    else:
      break

  save_dict_to_file(db)
```

```python
main()
```
